# Simulation of Coupled Harmonic Oscillators

Define a Python simulation _kernel_ which models the evolution of a 1-dimensional system of 3 masses linked by springs.  The masses are unconstrained.  Each call to `evolve` updates the forces, velocities, and positions of the 3 masses over a fixed, finite timestep defined by the `dt` instance variable, defaulting to 1 ms.  Damping can be introduced by a single velocity dependent damping factor, `damping`.

```python
import numpy as np
import pylab as plt
%matplotlib inline
```

```python
class sim_kernel(object):
    def __init__(self, x, M, k1=1.0, k2=1.0, damping=0.0):
        self.x = x
        self.v = np.zeros(3, 'd')
        self.M = M
        self.t = 0
        self.dt  = 0.001
        self.rec = 250
        self.n   = 0
        self.history = np.array(x)
        self.k1  = k2
        self.k2  = k2
        self.damping = damping
        
    def evolve(self):
        a = np.empty(3, 'd')
        a[0] = self.k1 * (self.x[1] - self.x[0])
        a[1] = (self.k1 * (self.x[0] - self.x[1]) + self.k2 * (self.x[2] - self.x[1]))
        a[2] = self.k2 * (self.x[1] - self.x[2])
        for i in range(3):
            a[i] -= self.damping * self.v[i]
            a[i] /= self.M[i]
        self.v += a * self.dt
        self.x += self.v * self.dt
        self.t += self.dt
        self.n += 1
        if self.n % self.rec == 0:
            self.history = np.vstack((self.history, np.array(self.x)))
```

Create the simulation kernel, giving masses with initial positions.

```python
k = sim_kernel(np.array([0, 0, 1], 'd'), np.array([1, 3, 2], 'd'))
for i in range(25000):
    k.evolve()
t = np.arange(101) * 0.25
plt.plot(t, k.history[:,0], label='Mass 1')
plt.plot(t, k.history[:,1], label='Mass 2')
plt.plot(t, k.history[:,2], label='Mass 3')
plt.xlabel('Time (sec)')
plt.ylabel('Position (m)')
```

Now, try with damping ...

```python
kd = sim_kernel(np.array([0, 0, 1], 'd'), np.array([1, 3, 2], 'd'), damping=0.25)
for i in range(25000):
    kd.evolve()
t = np.arange(101) * 0.25
plt.plot(t, kd.history[:,0], label='Mass 1')
plt.plot(t, kd.history[:,1], label='Mass 2')
plt.plot(t, kd.history[:,2], label='Mass 3')
plt.xlabel('Time (sec)')
plt.ylabel('Position (m)')
```
